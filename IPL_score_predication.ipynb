```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.linear_model import LinearRegression, Lasso,Ridge
from sklearn.ensemble import RandomForestRegressor

import warnings
warnings.filterwarnings("ignore")
```

```python
class score_prediction():
    def __init__(self,df):
        self.df=df
        
    def train_test(self):
        self.x= self.df.drop(['mid','date','venue','batsman','bowler','striker','non-striker','total'],axis=1)
        self.y=self.df['total']
        self.columns= list(self.x.columns)
        self.x_train,self.x_test,self.y_train,self.y_test=train_test_split(self.x,self.y,test_size=0.2,random_state=42)
    
    def lr_model(self):
        self.lr_model= LinearRegression()
        self.lr_model.fit(self.x_train,self.y_train)
        
    def evalution_metrics_for_train_data(self):
        y_pred=self.lr_model.predict(self.x_train)
        
        mae=mean_absolute_error(self.y_train,y_pred)
        print('MAE:',mae)
        print('*'*80)
        mse=mean_squared_error(self.y_train,y_pred)
        print('MSE:',mse)
        print('*'*80)
        r2=r2_score(self.y_train,y_pred)
        print('R2_Score:',r2)
        
    def evalution_metrics_for_test_data(self):
        y_pred_test=self.lr_model.predict(self.x_test)

        mae=mean_absolute_error(self.y_test,y_pred_test)
        print('MAE:',mae)
        print('*'*80)
        mse=mean_squared_error(self.y_test,y_pred_test)
        print('MSE:',mse)
        print('*'*80)
        r2=r2_score(self.y_test,y_pred_test)
        print('R2_Score:',r2)
    
    def hyperparameter_lasso_for_lr(self):
        lasso= Lasso()
        param= {'alpha':np.arange(0.01,3,0.01)}
        
        self.lr_rscv=RandomizedSearchCV(lasso,param,cv=5)
        self.lr_rscv.fit(self.x_train,self.y_train)

    def evalution_metrics_for_train_data_with_hypr_lasso(self):
        y_pred=self.lr_rscv.predict(self.x_train)
        print('Evalution Metrics For train data with Hyper parameter')
        mae=mean_absolute_error(self.y_train,y_pred)
        print('MAE:',mae)
        print('*'*80)
        mse=mean_squared_error(self.y_train,y_pred)
        print('MSE:',mse)
        print('*'*80)
        r2=r2_score(self.y_train,y_pred)
        print('R2_Score:',r2)
        
    def evalution_metrics_for_test_data_with_hyper_lasso(self):
        y_pred_test=self.lr_rscv.predict(self.x_test)
        print('Evalution Metrics For test data with Hyper parameter')
        mae=mean_absolute_error(self.y_test,y_pred_test)
        print('MAE:',mae)
        print('*'*80)
        mse=mean_squared_error(self.y_test,y_pred_test)
        print('MSE:',mse)
        print('*'*80)
        r2=r2_score(self.y_test,y_pred_test)
        print('R2_Score:',r2)

class random_forest():
    def __init__(self,):
        score_prediction.__init__(self,df)
        #self.df=df
    def train_test(self):
        self.x= self.df.drop(['mid','date','venue','batsman','bowler','striker','non-striker','total'],axis=1)
        self.y=self.df['total']
        self.columns= list(self.x.columns)
        self.x_train,self.x_test,self.y_train,self.y_test=train_test_split(self.x,self.y,test_size=0.2,random_state=12)
    
    def build_model(self):
        self.rf_model=RandomForestRegressor()
        self.rf_model.fit(self.x_train,self.y_train)
    
    def evalution_metrics_for_train_data_with_rf(self):
        y_pred=self.rf_model.predict(self.x_train)
        
        mae=mean_absolute_error(self.y_train,y_pred)
        print('MAE:',mae)
        print('*'*80)
        mse=mean_squared_error(self.y_train,y_pred)
        print('MSE:',mse)
        print('*'*80)
        r2=r2_score(self.y_train,y_pred)
        print('R2_Score:',r2) 
    def evalution_metrics_for_test_data_with_rf(self):
        y_pred_test=self.rf_model.predict(self.x_test)

        mae=mean_absolute_error(self.y_test,y_pred_test)
        print('MAE:',mae)
        print('*'*80)
        mse=mean_squared_error(self.y_test,y_pred_test)
        print('MSE:',mse)
        print('*'*80)
        r2=r2_score(self.y_test,y_pred_test)
        print('R2_Score:',r2)
    
    def hyperparameter_for_rf(self):
        rf_model= RandomForestRegressor(n_jobs=-1)
        param= {'n_estimators':np.arange(20),
            'criterion':['squared_error','absolute_error'],
            'max_depth':np.arange(3,8),
            'min_samples_split':np.arange(2,20),
            'min_samples_leaf':np.arange(2,20),
            'max_features':['sqrt'],
            'bootstrap':[True],
            'oob_score':[True]}
        self.rf_rscv=RandomizedSearchCV(rf_model,param,cv=5)
        self.rf_rscv.fit(self.x_train,self.y_train)
    
    def evalution_train_rf_hyper(self):
        y_pred=self.rf_rscv.predict(self.x_train)
        print("Evalution Metrics for Training data with Random Forest Algoritham")
        mae=mean_absolute_error(self.y_train,y_pred)
        print('MAE:',mae)
        print('*'*80)
        mse=mean_squared_error(self.y_train,y_pred)
        print('MSE:',mse)
        print('*'*80)
        r2=r2_score(self.y_train,y_pred)
        print('R2_Score:',r2) 
    
    def evalution_test_rf_hyper(self):
        y_pred_test=self.rf_rscv.predict(self.x_test)
        print("Evalution Metrics for Testing data with Random Forest Algoritham")
        mae=mean_absolute_error(self.y_test,y_pred_test)
        print('MAE:',mae)
        print('*'*80)
        mse=mean_squared_error(self.y_test,y_pred_test)
        print('MSE:',mse)
        print('*'*80)
        r2=r2_score(self.y_test,y_pred_test)
        print('R2_Score:',r2)
```
